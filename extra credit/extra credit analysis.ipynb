## Analysis

Multivariate regression

```python
import numpy as np   
import pandas as pd  
import matplotlib.pyplot as plt  

import statsmodels.formula.api as smf
%pylab inline
```

```python
taxi_14_1 = pd.read_csv('taxi_2014_1_data_boroct.csv')
```

```python
taxi_14_1.shape
```

```python
taxi_14_1.columns
```

```python
lm_14_1= smf.ols(formula = 'count ~ income + age + safety + sub_distances', data = taxi_14_1).fit()
print(lm_14_1.summary())
```

```python
taxi_14_1['count'][~taxi_14_1.safety.isnull()].shape
```

```python
plt.figure()
lmy_14_1=lm_14_1.predict()
plt.plot(lmy_14_1, lmy_14_1, 'b-',  label = 'predicted line')
plt.plot(lmy_14_1,taxi_14_1['count'][~taxi_14_1.safety.isnull()],'or', label = 'taxi 14-1')
plt.xlabel('Prediction', fontsize = 16)
plt.ylabel('Observation', fontsize = 16)
plt.title('predicting from the distribution of pickups for taxi 14-01')
plt.legend()
```
